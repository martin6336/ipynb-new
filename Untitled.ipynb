```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec

sns.set(style="whitegrid", palette="muted", color_codes=True)  
```

```python
x1=[1,2,3,4]
y1=[0.746239,0.746445,0.728835,0.733227]
x2=[5,10,15,20]
y2=[0.745973,0.746239,0.747411,0.747522]
x3=[0.1,0.2,0.4,0.6,0.8,0.1]
y3=[0.7462,0.747063,0.74515,0.740281,0.738199,0.73564]
x4=['avg','max','filter']
y4_1=[0.733137,0.736795,0.747063]
y4_2=[0.145184,0.15251,0.132595]

f, axes=plt.subplots(1, 2)
sns.pointplot(x=x1,y=y1,ax=axes[0])


sns.pointplot(x=x2,y=y2,ax=axes[1])


xlabel=['number of layers','dimension of embedding','learning rate/e-3','zero']
for i, ax in enumerate(axes.flat):
    ax.set(xlabel=xlabel[i], ylabel='AUC')
for ax in axes.flat:
    ax.label_outer()
plt.savefig('layer_embedding.pdf')
plt.show()
```

```python
f, axes=plt.subplots(1, 2)
sns.pointplot(x=x4,y=y4_1,ax=axes[0])

sns.pointplot(x=x4,y=y4_2,ax=axes[1])


xlabel=['type of filter','type of filter']
for i, ax in enumerate(axes.flat):
    ax.set(xlabel=xlabel[i], ylabel='AUC')
for ax in axes.flat:
    ax.label_outer()
plt.savefig('conv_type.pdf')
plt.show()

# plt.xlabel('learning rate')
# plt.ylabel('AUC')
# plt.tight_layout()
```
